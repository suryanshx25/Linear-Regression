```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['placement_exam_marks'])
plt.xlabel('CGPA')
plt.ylabel('Placement Exam Marks')
```

```python
X = df.iloc[:,0:1]
Y = df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

```

X and Y: These are the datasets you want to split. Typically, X would represent the feature data (input variables) and Y would represent the corresponding target data (output or labels).

train_test_split: This is a function provided by scikit-learn that allows you to split your dataset into training and testing subsets. It's commonly used to evaluate the performance of machine learning models.

test_size=0.2: This parameter specifies the proportion of the dataset that should be allocated for testing. In this case, 0.2 means that 20% of the data will be used for testing, and the remaining 80% will be used for training.

random_state=2: This parameter sets the random seed for the random number generator that's used during the splitting process. Setting a random seed ensures that the random split is reproducible. In other words, if you use the same random seed in different runs, you will get the same split each time. This can be helpful for consistent results when debugging or sharing code.

```python
# X_train: This variable will hold the training feature data after the split.
# X_test: This variable will hold the testing feature data after the split.
# Y_train: This variable will hold the training target data (labels) after the split.
# Y_test: This variable will hold the testing target data (labels) after the split.

# Assuming you have defined X and Y

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
X_test
```

```python
Y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
lr.predict(X_test.iloc[2].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['placement_exam_marks'])
plt.plot(X_train,lr.predict(X_train),color = 'red')
plt.xlabel('CGPA')
plt.ylabel('Placement Exam Marks')
```
